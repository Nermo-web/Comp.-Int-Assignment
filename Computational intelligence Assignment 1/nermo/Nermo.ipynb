## Table

```python
##################--- Loading dataset ---############################
df_acc = pd.read_csv("./acc.csv")
df_labels = pd.read_csv("./labels.csv")
df_mag = pd.read_csv("./mag.csv")
df_velo = pd.read_csv("./velo.csv")
df_orin = pd.read_csv("./orin.csv")

##################--- Transforming the timestamp ---##########################

#############################################################
df_acc = pd.read_csv("./acc.csv")
df_acc["Timestamp"] = pd.to_datetime(df_acc["Timestamp"])
origine_time = df_acc["Timestamp"].loc[0]
df_acc["Timestamp"] = (df_acc["Timestamp"]-origine_time).dt.total_seconds()
#############################################################
df_orin = pd.read_csv("./orin.csv")
df_orin["Timestamp"] = pd.to_datetime(df_orin["Timestamp"])
origine_time = df_orin["Timestamp"].loc[0]
df_orin["Timestamp"] = (df_orin["Timestamp"]-origine_time).dt.total_seconds()
#############################################################
df_velo = pd.read_csv("./velo.csv")
df_velo["Timestamp"] = pd.to_datetime(df_velo["Timestamp"])
origine_time = df_velo["Timestamp"].loc[0]
df_velo["Timestamp"] = (df_velo["Timestamp"]-origine_time).dt.total_seconds()
#############################################################
df_mag = pd.read_csv("./mag.csv")
df_mag["Timestamp"] = pd.to_datetime(df_mag["Timestamp"])
origine_time = df_mag["Timestamp"].loc[0]
df_mag["Timestamp"] = (df_mag["Timestamp"]-origine_time).dt.total_seconds()
#############################################################
```

## Pre-processing 

### Standarization

Standardization is a transform for data with a Gaussian distribution.

It subtracts the mean and divides the result by the standard deviation of the data sample. This has the effect of transforming the data to have mean of zero, or centered, with a standard deviation of 1. This resulting distribution is called a standard Gaussian distribution, or a standard normal, hence the name of the transform.

```python
from sklearn.preprocessing import StandardScaler
from numpy import array

timestamp = df_acc["Timestamp"]
table_stand = StandardScaler().fit_transform(table[['Acc_X', 'Acc_Y', 'Acc_Z', 
        'AngV_X', 'AngV_Y', 'AngV_Z',
       'Mag_X', 'Mag_Y', 'Mag_Z',
       'Orin_X', 'Orin_Y', 'Orin_Z']])
```

```python
table_stand = np.concatenate((timestamp.to_numpy().reshape(-1,1),table_stand),axis=1)
```

```python
features = range(12)
features_name = ['Acc_X', 'Acc_Y', 'Acc_Z', 
        'AngV_X', 'AngV_Y', 'AngV_Z',
       'Mag_X', 'Mag_Y', 'Mag_Z',
       'Orin_X', 'Orin_Y', 'Orin_Z']

for i in features :
    
    plt.figure(figsize=(14, 5))
    plt.title(features_name[i])
    plt.plot(table_stand[:,0],table_stand[:,i+1],label="with standarization")
    plt.plot(table_stand[:,0],table_numpy[:,i+1],label="raw data")
    plt.title(features_name[i])
    plt.legend()
    plt.show()
```

## Label Encoding

## Split train/test/validation

**Split tain 60 % / test 20 % / validation 20 %**

```python
from sklearn.model_selection import train_test_split
```

```python
Train, Test = train_test_split(table_selected_activities, test_size=0.2, random_state=1)
Train, Validation = train_test_split(Train, test_size=0.25, random_state=1)
```

```python
Train.head()
```

```python
Counter(list(zip(Train.Activity)))
```

```python
Counter(list(zip(Test.Activity)))
```

## Standarization

```python
###################--- Train ---####################################


Train_X = StandardScaler().fit_transform(Train[['Acc_X', 'Acc_Y', 'Acc_Z', 
        'AngV_X', 'AngV_Y', 'AngV_Z',
       'Mag_X', 'Mag_Y', 'Mag_Z',
       'Orin_X', 'Orin_Y', 'Orin_Z']])
Train_Y = Train[['Activity']]

###################--- Test ---####################################

Test_X = StandardScaler().fit_transform(Test[['Acc_X', 'Acc_Y', 'Acc_Z', 
        'AngV_X', 'AngV_Y', 'AngV_Z',
       'Mag_X', 'Mag_Y', 'Mag_Z',
       'Orin_X', 'Orin_Y', 'Orin_Z']])
Test_Y = Test[['Activity']]

###################--- Validation ---####################################

Validation_X = StandardScaler().fit_transform(Validation[['Acc_X', 'Acc_Y', 'Acc_Z', 
        'AngV_X', 'AngV_Y', 'AngV_Z',
       'Mag_X', 'Mag_Y', 'Mag_Z',
       'Orin_X', 'Orin_Y', 'Orin_Z']])
Validation_Y = Validation[['Activity']]
```

```python
Train_X.shape,Train_Y.shape,Test_X.shape,Test_Y.shape
```

## Hyperparameters

```python
num_hidden_units_1 = 100
num_hidden_units_2 = 100
num_percentage_dropout = 0.5
batch_size = 32
epochs = 128
```

## Model

```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
n_features = 12
n_outputs = 4

model = tf.keras.Sequential()

model.add(layers.Dense(num_hidden_units_1, input_shape=(n_features,)))
model.add(layers.Dropout(num_percentage_dropout))
model.add(layers.Dense(num_hidden_units_2, activation='relu'))
model.add(layers.Dense(n_outputs, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Plot the model

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
model.summary()
```

### Train the model

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
history = model.fit(Train_X, Train_Y, epochs=epochs, batch_size=batch_size,validation_data=(Test_X, Test_Y),callbacks=[callback])
```

```python
print("Evaluate on Validation data")
results = model.evaluate(Validation_X, Validation_Y, batch_size=batch_size)
print("Validation loss, Validation accuracy:", results)
```

## History : Accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

## History : Loss

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

## Confusion Matrix on Validation data

```python
y_pred = model.predict(Validation_X)
y_pred = np.argmax(y_pred, axis=1)

y_true = Validation_Y
```

```python
confusion_mtx = tf.math.confusion_matrix(y_true, y_pred)
features = ['Walking', 'Vibration','Standing','Sitting']
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_mtx,
            xticklabels=features,
            yticklabels=features,
            annot=True, fmt='g')
plt.xlabel('Prediction')
plt.ylabel('Label')
plt.show()
```
